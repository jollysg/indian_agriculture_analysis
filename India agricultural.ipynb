```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data_file_path = '5-Year Foodgrain 2016-17 to 2020-21.xlsx'
agr_df = pd.read_excel(data_file_path, sheet_name=None)
```

```python
grains = df['All India Estimates of Area, Production and Yield of Foodgrains']
total_prod = df['Unnamed: 2']

px.scatter(x=grains, y=total_prod)
```
